TITANIC DATASET

```python
df=pd.read_csv("titanic.csv")
df
```

```python
sns.countplot(x="Survived",hue="Sex",data=df)
```

```python
sns.distplot(df['Age'].dropna(),kde=True)
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
df['Age'].hist(alpha=0.5)
```

```python
df.isnull().sum()
```

```python
df.Age=df.Age.fillna(df.Age.mean())
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
v=['Cabin']
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
df[v]=imputer.fit_transform(df[v])
df
```

```python
df.isnull().sum()
```

```python
df=df.fillna(df.Embarked.value_counts().index[0])
```

```python
df.isnull().sum()
```

```python
df
```

```python
df=df.drop(['Name','Ticket','Fare','Cabin','Embarked'],axis=1)
df
```

```python
df.Sex=df.Sex.map({'male':0,'female':1})
```

```python
df
```

```python
df=df.drop(['Pclass','SibSp','Parch'],axis=1)
df
```

```python
X=df[['PassengerId','Sex','Age']]
Y=df['Survived']
```

```python
X
```

```python
Y
```
